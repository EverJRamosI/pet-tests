# Usage PetPropy

## Imports

```python
import petpropy as pp
import numpy as np
import matplotlib.pyplot as plt
```

## Input Data for Factor Compressibility Gas

```python
pressures_array = np.arange(500, 10500, 500) #psia
pressures_list = list(range(500, 10500, 500)) # psia
temperature = 654 #oR
pres_critica = 719.565742 #psia
temp_critica = 548.401303 #psia
```

### Usage z_g and functions

```python
z = pp.z_g.dranchuk_abou_kassem(pressures_array, temperature, pres_critica, temp_critica)
print(z)
```

```python
plt.plot(pressures_array, z, color='red', linestyle='--')
plt.ylabel('z')
plt.xlabel('Pressures')
plt.grid(True)
plt.show()
```

```python
z = pp.z_g.dranchuk_purvis_robinson(pressures_list, temperature, pres_critica, temp_critica)
print(z)
```

```python
plt.plot(pressures_array, z, color='green', linestyle='--')
plt.ylabel('z')
plt.xlabel('Pressures')
plt.grid(True)
plt.show()
```

## Input Data for Solution Gas-Oil Ratio

```python
pressures = np.arange(500, 10500, 500) #psia
temperature = 654 #oR
sp_g = 0.8 #specific gravity gas
api = 35 # API
pressure_b = 2500 #Bubble pressure
```

### Usage Rs and Functions

```python
rs = pp.Rs.almarhoun(pressures, temperature, sp_g, api, Pb=pressure_b)
print(rs)
```

```python
plt.plot(pressures, rs, color='green', linestyle='--')
plt.ylabel('Rs (scf/stb)')
plt.xlabel('Pressures (psia)')
plt.grid(True)
plt.show()
```

### Usage mu_o and Functions

```python
v = pp.mu_o.mu_oil(pressures, temperature, api, rs, Pb=pressure_b)
print(v)
```

```python
plt.plot(pressures, v, color='red', linestyle='--')
plt.ylabel('Viscosity (cp)')
plt.xlabel('Pressures (psia)')
plt.grid(True)
plt.show()
```
